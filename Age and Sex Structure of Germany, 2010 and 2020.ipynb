```python
import pandas as pd
```

## 1. Year 2020

### 1) Data Definition

```python
# Men's data
file_name = '12411-0006_2020.xlsx'
df_m = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A:B')
df_m.dropna(axis = 'index', inplace = True)
df_m
```

```python
# Women's data
df_w = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A,C')
df_w.dropna(axis = 'index', inplace = True)
df_w
```

### 2) Data Visualisation

```python
import matplotlib.pyplot as plt
```

```python
male = df_m.columns[0]
female = df_w.columns[0]

plt.figure(figsize = (7, 18))
plt.xlim(-1700, 1700)
plt.barh(df_m.index, -df_m[male] // 1000, label = 'male') # Unit: K
plt.barh(df_w.index, df_w[female] // 1000, label = 'female') # Unit: K
plt.legend(loc = 'upper center', ncol = 2)
plt.grid(axis = 'x', alpha = 0.5)
plt.title("Germany's Population Pyramid of 2020")
plt.savefig('Population Pyramid of 2020.png', dpi = 50)
plt.show()
```

## 3. Comparison betweeen 2010 and 2020

```python
fig, axs = plt.subplots(1, 2, figsize = (14, 15), sharey='all')
fig.suptitle('Age and Sex Structure of Germany, 2010 and 2020')

# 2010 Graph
## Male
file_name = '12411-0006_2010.xlsx'
df_m = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A:B')
df_m.dropna(axis = 'index', inplace = True)

## Female
df_w = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A,C')
df_w.dropna(axis = 'index', inplace = True)

axs[0].barh(df_m.index, -df_m[male] // 1000, label = 'male') # Unit: K
axs[0].barh(df_w.index, df_w[female] // 1000, label = 'female') # Unit: K
axs[0].set_title("Germany's Population Pyramid of 2010")
axs[0].legend(loc = 'upper center', ncol = 2)
axs[0].set(xlabel = 'Population in K', ylabel = 'Age') 
axs[0].grid(axis = 'x', alpha = 0.5)
axs[0].set_xlim(-1700, 1700)


# 2020 Graph
## Male
file_name = '12411-0006_2020.xlsx'
df_m = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A:B')
df_m.dropna(axis = 'index', inplace = True)

## Female
df_w = pd.read_excel(file_name, skiprows = 5, nrows = 87, index_col = 'Altersjahre', usecols = 'A,C')
df_w.dropna(axis = 'index', inplace = True)

axs[1].barh(df_m.index, -df_m[male] // 1000, label = 'male') # Unit: K
axs[1].barh(df_w.index, df_w[female] // 1000, label = 'female') # Unit: K
axs[1].set_title("Germany's Population Pyramid of 2020")
axs[1].legend(loc = 'upper center', ncol = 2)
axs[1].set(xlabel = 'Population in K', ylabel = 'Age') 
axs[1].grid(axis = 'x', alpha = 0.5)
axs[1].set_xlim(-1700, 1700)

plt.show()
```
